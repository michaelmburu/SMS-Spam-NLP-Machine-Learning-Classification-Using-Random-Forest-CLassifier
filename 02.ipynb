```python
import nltk
import pandas as pd
import string
import re
ps = nltk.PorterStemmer()

pd.set_option('display.max_colwidth', 200)
data = pd.read_csv("SMSSpamCollection.tsv", sep='\t', header=None)
data.columns = ['label', 'body_text']
data.head()
```

```python
print("Input data has {} rows and {} columns".format(len(data),
                                                     len(data.columns)))
```

```python
#How many are spam/ham are there?
print("Out of {}, {} are spam, {} are ham".format(len(data), 
                                                   len(data[data['label']=='spam']), 
                                                   len(data[data['label'] == 'ham'])
                                                  ))
```

```python
#How much missing data is there?

print("Number of null in label: {}".format(data['label'].isnull().sum()))
print("Number of null in text: {}".format(data['body_text'].isnull().sum()))
```

```python
#Removing Punctaution
import string

def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct

data['body_text_clean'] = data['body_text'].apply(lambda x: remove_punct(x))

data.head()
```

```python
#Tokenizing

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

data['body_text_tokenized'] = data['body_text_clean'].apply(lambda x: tokenize(x.lower()))

data.head()
```

```python
#Removing stopwords


stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text


data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

data.head()
```
